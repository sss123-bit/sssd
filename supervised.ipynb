```python
s=MinMaxScaler() 
ndf=s.fit_transform(df)
```

```python
ndf=pd.DataFrame(ndf,columns=['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
       'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
       'Proanthocyanins', 'Color intensity', 'Hue',
       'OD280/OD315 of diluted wines', 'Proline', 'Wine Type'])
```

```python
ndf
```

# here we can se data is normalized by using min max
# here every data is in range [0,1] 


```python
ndf=ndf.drop(labels=['Wine Type'],axis=1)
```

```python
ndf
```

```python
ndf['Wine Type']=df['Wine Type']
```

```python
ndf
```

#   here we normalise everything except our label/target as it will effect our model and will throw an error as unknown label 
# Hence whe keep the Label un normalised

```python
from sklearn.model_selection import train_test_split
X=ndf[['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
       'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
       'Proanthocyanins', 'Color intensity', 'Hue',
       'OD280/OD315 of diluted wines', 'Proline']] 
y=ndf['Wine Type']
trainx,testx,trainy,testy=train_test_split(X,y,test_size=0.2,random_state=2)
```

# Spliting the data for  training and  testing 

```python
from sklearn.decomposition import PCA 
pca=PCA(0.95) 
trainx1=pca.fit_transform(trainx)
testx1=pca.transform(testx)
```

```python
ev=pca.explained_variance_ratio_
```

```python
ev
```

# HERE we can see the varience of data set attributes vectors in plane 

# ------------------------------------------------x---------------------------------------------------------

### Let us train and test our model using  
# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression 
modelo=LogisticRegression()
modelo.fit(trainx1,trainy)
modelo.score(testx1,testy)
```

```python
from sklearn.metrics import confusion_matrix
predy=modelo.predict(testx1)
```

```python
cml=confusion_matrix(testy,predy)
```

```python
cml=pd.DataFrame(cml)
```

```python
cml
```

# from the confusion matrix we can see that only 1 wine type was predicted wrong 


# here from confusion matrix we can conclude that the wine type which was supposed to be label '0' was preicted as label '1' which is an  error
