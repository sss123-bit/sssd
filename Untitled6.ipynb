```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import warnings
warnings.filterwarnings(action='ignore')
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
df = pd.read_csv('indian_liver_patient_dataset.csv')
```

```python
df
```

```python
scaler = MinMaxScaler()
features=['age','TB','DB','alkphos','sgpt','sgot','TP','ALB','A_G','class'] 
scaled_values=scaler.fit_transform(df[features])
df["norm_age"]=scaled_values[:,0]
df["norm_TB"]=scaled_values[:,1]
df["norm_DB"]=scaled_values[:,2]
df["norm_sgpt"]=scaled_values[:,3]
df["norm_TP"]=scaled_values[:,4]
df["norm_ALB"]=scaled_values[:,5]
df["norm_alkphos"]=scaled_values[:,6]
df["norm_sgot"]=scaled_values[:,7]
df["norm_A_G"]=scaled_values[:,8] 
df['norm_class']=scaled_values[:,9]
```

```python
df
```

```python
df.corr(method='spearman')
```

```python
df.var()
```

# Logistic 

```python
model = LogisticRegression()
X =df[['norm_age','norm_DB','norm_sgpt','norm_alkphos']]
y=df['norm_class']
trainx,testx,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
y_train.shape,y_test.shape 
model.fit(trainx,y_train)

print(model.score(trainx,y_train),model.score(testx,y_test))
```

# decision trees

```python
model=DecisionTreeClassifier(min_samples_split=10)
X =df[['norm_age','norm_TB','norm_sgpt','norm_alkphos']]
y=df['norm_class']

trainx,testx,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
y_train.shape,y_test.shape
model.fit(trainx,y_train)
model.score(testx,y_test)
```

# knn

```python
model = KNeighborsClassifier(n_neighbors=3,metric="euclidean")
X =df[['norm_age','norm_TB','norm_sgpt','norm_alkphos']]
y=df['norm_class']
trainx,testx,y_train,y_test=train_test_split(X,y,test_size=0.2)
model.fit(trainx,y_train)

train_accuracy = model.score(trainx,y_train)
test_accuracy = model.score(testx,y_test)
print(train_accuracy,test_accuracy)
```

## svm

```python
model=SVC(kernel='rbf')
X =df[['norm_age','norm_TB','norm_sgpt','norm_alkphos']]
y=df['norm_class']
trainx,testx,y_train,y_test=train_test_split(X,y,test_size=0.2)
model.fit(trainx,y_train)
train_accuracy = model.score(trainx,y_train)
test_accuracy = model.score(testx,y_test)
print(train_accuracy,test_accuracy)
```
