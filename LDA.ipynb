```python
le_iris=LabelEncoder()
le_iris=le_iris.fit(iris['iris'])
iris['iris']=le_iris.transform(iris['iris'])
```

```python
X=iris.loc[:,['sepal length','sepal width','petal length','petal width']]
y=iris.loc[:,['iris']]
```

```python
scaler=MinMaxScaler()
nX=pd.DataFrame(scaler.fit_transform(X))
```

```python
nX.columns=X.columns
```

```python
nX
```

```python
iris
```

```python
y
```

```python
trainnX,testnX,trainy,testy=train_test_split(nX,y,test_size=0.2,random_state=4)
```

```python
lda=LDA(n_components=3)
trainnX=lda.fit_transform(trainnX,trainy)
testnX=lda.transform(testnX)
```

```python
dec=DecisionTreeClassifier()
model=dec.fit(trainnX,trainy)
```

```python
preds=model.predict(testnX)
```

```python
preds
```

```python
model.score(trainnX,trainy)
```

```python
cm=confusion_matrix(testy,preds)
print(cm)
```
