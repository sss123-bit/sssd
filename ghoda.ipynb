```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('horse.csv')
```

```python
df
```

```python
df['pulse'].isna().sum()
```

```python
for i in df.columns:
    if df[i].dtype in ['int64','float64'] and df[i].isna().sum()>0:
        df[i].fillna(value=df[i].mean(),inplace=True)
    elif df[i].dtype not in ['int64','float64'] and df[i].isna().sum()>0:  
        df[i].fillna(value=df[i].mode()[0],inplace=True)
```

```python
df['rectal_temp'].fillna(value=df['rectal_temp'].mean(),inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

```python
k=LabelEncoder()
for i in df.columns:
    if df[i].dtype not in ['int64','float64']:
        df[i]=k.fit_transform(df[i])
```

```python
df
```

```python
df.columns
```

```python
s=MinMaxScaler()
```

```python
X=s.fit_transform(df.drop(columns=['outcome']))
```

```python
x=pd.DataFrame(X,columns=(['surgery', 'age', 'hospital_number', 'rectal_temp', 'pulse',
       'respiratory_rate', 'temp_of_extremities', 'peripheral_pulse',
       'mucous_membrane', 'capillary_refill_time', 'pain', 'peristalsis',
       'abdominal_distention', 'nasogastric_tube', 'nasogastric_reflux',
       'nasogastric_reflux_ph', 'rectal_exam_feces', 'abdomen',
       'packed_cell_volume', 'total_protein', 'abdomo_appearance',
       'abdomo_protein', 'surgical_lesion', 'lesion_1', 'lesion_2',
       'lesion_3', 'cp_data']))
```

```python
x
```

```python
features=x.columns
target=['outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x[features],df[target],test_size=0.2,random_state=100)
```

```python

from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(n_estimators=10)
model.fit(x_train,y_train)
train_accuracy = model.score(x_train,y_train)
test_accuracy = model.score(x_test,y_test)
print(train_accuracy,test_accuracy)
```

```python
feature_imps = pd.DataFrame(np.array([features,
                                      model.feature_importances_]).T,
                            columns=["feature","importance"])
feature_imps.sort_values(by="importance",ascending=False)
```

```python
org_feat=['surgery','total_protein','temp_of_extremities','lesion_1','abdomo_protein','packed_cell_volume','nasogastric_tube','rectal_temp']
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=8)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
 pca.explained_variance_ratio_
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=8,
                               min_samples_split=20,
                               min_impurity_decrease=0.05)
model.fit(x_train,y_train)
train_accuracy = model.score(x_train,y_train)
test_accuracy = model.score(x_test,y_test)
print(train_accuracy,test_accuracy)
```
